```python
#MAKE A HUT
import matplotlib.pyplot as plt

x1=[ 2, 2, 2.50,2.50, 2 ]
y1= [2, 4, 4, 2, 2 ]
x2=[2.2, 2.2, 2.1, 2 ]
y2=[2, 4, 6, 4]
x3=[2.5, 2.45, 2.1]
y3=[4, 6 , 6]
x4=[2.15, 2.15, 2.05, 2.05]
y4=[2, 3, 3 , 2.05 ]
x5=[2.25, 2.25, 2.45, 2.45, 2.25]
y5=[2.5, 3.5, 3.5, 2.5, 2.5]
x6=[2.35,2.35 ]
y6=[2.5, 3.5]
x7=[2.25, 2.45]
y7=[3, 3]


plt.title("HUT")
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
plt.plot(x5,y5)
plt.plot(x6,y6)
plt.plot(x7,y7)
plt.savefig("hut.jpg")
plt.show()
```
