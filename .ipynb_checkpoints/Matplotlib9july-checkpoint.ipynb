```python
import matplotlib.pyplot as plt
x=["Monday", "Tuesday", "Wednesday","Thursday", "Friday", "Saturday", "Sunday"]
y=[11,23,67,89,24,25,71]

plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Graph 1")
plt.plot(x,y)
plt.show()
```

```python
# MAKE A SMILEY
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

Circle=Circle([3.5,9], .5, fill=False)

plt.axis('equal')
plt.gca(). add_patch(Circle)
plt.plot()

plt.show()
```
