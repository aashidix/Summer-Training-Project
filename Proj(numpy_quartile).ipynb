```python
import numpy as np
```

```python
from  PIL import Image
import matplotlib.pyplot as plt
```

```python
img= Image.open(r"/Users/Aashi_Dixit/Desktop/SUMMER TRAINING/dragon.jpg") #save img in same folder as the file we're working on; can also give path but it can lead to errors
img=np.array(img)

h, w = img.shape[:2]
q1, q2, q3, q4 = img[:h//2,:w//2], img[:h//2,w//2:], img[h//2:,:w//2], img[h//2:,w//2:]

q1[:,:,0] = np.clip(q1[:,:,0] + 50, 0, 255)   # Increase Red (top-left)
q2[:,:,1] = np.clip(q2[:,:,1] + 50, 0, 255)   # Increase Green (top-right)
q3[:,:,2] = np.clip(q3[:,:,2] + 50, 0, 255)   # Increase Blue (bottom-left)
q4[:] = 255 - q4  
plt.imshow(img)
plt.axis("on")
plt.show()
```
